```python
import pandas as pd
import numpy as np
import time
from matplotlib import pyplot as plt
from matplotlib import dates as dates

df = pd.read_parquet("training/data/nfcore-small-20260623-batch-cleaned.parquet")

df = df.set_index("_time")
print(df.columns)
df = df["interval_energy"]

range=300
time_frame_np = np.arange(df.index.values[0], df.index.values[-1], 1, dtype='datetime64[s]')

start = int(time_frame_np.size / 2 - range / 2)
end = int(time_frame_np.size / 2 + range / 2)

_ , ax = plt.subplots(figsize=(10, 5))

ax.plot(
    time_frame_np[start:end],
    df[start:end],
    label="Actual Energy",
    linewidth=1.0,
)

ax.set_xlabel("Time", fontsize=10, labelpad=4)
ax.xaxis.set_major_locator(dates.SecondLocator(interval=30))
ax.xaxis.set_major_formatter(dates.DateFormatter('%H:%M:%S'))
ax.set_ylabel("Interval Energy (Wh)", fontsize=12, labelpad=4)
ax.tick_params(axis="both", labelsize=10)
ax.legend(
    loc="upper right",
    fontsize=10,
    frameon=True
)
ax.set_title("Energy of Dataset", fontsize=14)

plt.savefig(f"training/plots/actual_vs_predicted_interval_energy-{time.strftime("%m%d%H%M%S")}.png", bbox_inches="tight", dpi=300)
plt.close()
```
